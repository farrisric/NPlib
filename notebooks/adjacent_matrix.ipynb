```python
from ase.cluster import Octahedron
from npl.core import Nanoparticle as NP
from scipy.spatial import KDTree


system = Octahedron('Pt',5,1)
for i in range(2):
    system[i].symbol = 'Pd'
system = NP.from_atoms(system)

tree1 = KDTree(system.get_positions())
tree2 = KDTree(system.get_positions())
```

```python
dmat = tree1.query_ball_tree(tree2, 3)
dmat
```
