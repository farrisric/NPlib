```python
import sys
from ase.cluster import Octahedron
from npl.core import Nanoparticle as NP

system = Octahedron('Pt',5,1)
for i in range(2):
    system[i].symbol = 'Pd'
system = NP.from_atoms(system)
conn_m = system.get_connectivity_matrix()
occupation_a, occupation_b = system.get_symbols_lists()
```
