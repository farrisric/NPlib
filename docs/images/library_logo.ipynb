```python
from ase.visualize.plot import plot_atoms
from npl.core import Nanoparticle
import matplotlib.pyplot as plt  

plt.rcParams['figure.dpi'] = 100

atoms = Nanoparticle()
atoms.truncated_octahedron(8, 1, {'Au': 0.5, 'Pd': 0.5})

atoms = atoms.get_ase_atoms()
indices_r = [x.index for x in atoms if x.position[2] > 0]
indices_l = [x.index for x in atoms if x.position[2] < 0]

for i in atoms:
    if i.index in indices_r:
        i.symbol = 'Zn'
    else:
        i.symbol = 'Pd'

plot_atoms(atoms, rotation='0x,90y')
plt.axis('off')
```
